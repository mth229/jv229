<p>Let $f$ be a function defined on the real line. M. Stemlund proved in 2001 (The College Mathematics Journal, 32, 194-196) the following:</p>

<blockquote>
<p>The function $f(x)$ is a quadratic polynomial if and only if for any $a$ and $b$ the  intersection point of the tangents lines to  $f(x)$ at $(a, f(a))$  and $(b, f(b))$ happens at $x=(a+b)/2$.</p>
</blockquote>

<p>We wish to investigate this. Here are some tools for our toolbelt:</p>

```julia
using Plots
using Roots
Base.ctranspose(f::Function) = D(f)
tangent(f, c) = x -> f(c) + f'(c)*(x-c)
```

<p>We can graph two tangent lines using this pattern:</p>

```julia
f(x) = exp(x)
a,b = 1, 2
plot(f, 1, 2)
plot!(tangent(f, a))
plot!(tangent(f, b))
```

<p>Here we can see the intersection point is <strong>not</strong> at <code>&#40;a&#43;b&#41;/2 &#61; 1.5</code>, but a bit to the right of that.</p>
