<p>But there is more to functions...</p>

<h4>Julia functions can have more than one variable</h4>

```julia
rectarea(b, h) = b * h		# area of rectangle
```

<p>The <code>log</code> function is an example, where <code>log&#40;b,x&#41;</code> will find the log base $b$ of $x$.</p>

<h4>Julia functions with different signatures, can have the same name!</h4>

```julia
rectarea(b) = rectarea(b,b)	# area of square using area of rectangle
```

<p>Which function is used depends on the arguments that are passed when calling the function. Hence, <code>log&#40;x&#41;</code> will find log base $e$ of $x$ whereas <code>log&#40;10,x&#41;</code> can be used to find log base 10.</p>

<h4>Multiline functions are defined with the function keyword</h4>

```julia
function multiline_function(x)
	 top = x^2 - 2x
	 bottom = x^4 - 4x
	 top/bottom		# last line is return value
end				# don't forget the end
```

<h4>Functions can have keyword arguments (like parameters). These have default values:</h4>

```julia
mxplusb(x; m=0, b=1) = m*x + b
mxplusb(10, m=3, b=4)		# arguments are named
```

<p>The semicolon separates the two types of arugments.</p>

<h4>Functions can be used as arguments to other functions:</h4>

```julia
using Plots	    		# for plotting
gadfly()
plot(sin, 0, 2pi)		# pass function for first argument plot(f,a,b)
```

<p>Using functions this way will form a <em>very</em> common pattern.</p>

<h4>Functions can be anonymous.</h4>

<p>These are defined via the pattern: "<code>args -&gt; body</code>". Look for the arrow <code>-&gt;</code>:</p>

```julia
plot(x -> sin(x)*cos(2x), 0, 2pi)
```

<h2>Cases</h2>

<p>Some functions are defined in terms of cases. For example, a cell phone plan might depend on the data used through:</p>

<blockquote>
<p>The amount is 40 dollars for the first 1 Gb  of data, and 10 dollars more for each <em>additional</em> Gb of data.</p>
</blockquote>

<p>This function has two cases to consider: one if the data is less than 1 Gb and the other when it is more.</p>

<p>How to write this in <code>julia</code>?</p>

<p>The ternary operator <code>predicate ? expression1 : expression2</code> has three pieces: a predicate question, such as <code>x &lt; 10</code> and two expressions, the first is evaluated if the predicate is <code>true</code> and the second if the predicate is <code>false</code>. They are useful to write functions that are defined by cases. (They are a light-weight form of the traditional <code>if-then-else</code> construct.)</p>

<p>The example above, could then be done with:</p>

```julia
f(d) = d <= 1.0 ? 40.00 : 40.00 + 10.00 * (d - 1.0)
```

<h2>Returning a function</h2>

<p>Familiar mathematical functions take a real number and return a real number. However, the concept of a function is more general. With <code>julia</code> it is useful to write functions that take functions as arguments, and return a derived function as an output. For the return value an <em>anonymous function</em> is typically used. These are specified as <code>arg -&gt; rule</code>, where the arg is a variable, like <code>x</code>, and the rule can depend on the value of <code>x</code>.</p>

```julia
function secant(f, a, b)
  m = (f(b) - f(a)) / (b - a)
  x -> f(a) + m * (x - a)
end
```
